Question 1
plot a line plot between a and b:
a=np.arange(40,50)
b=np.arange(50,60)


```python
import numpy as np
import matplotlib.pyplot as plt
a = np.arange(40,50)
b = np.arange(50,60)
plt.plot(a,b,"r--",linewidth = 5)
plt.xlabel("X - Axis")
plt.ylabel("Y - Axis")
plt.title("Line Plot")
plt.show()
```

Question 2
Plot a line plot showing the sales trend in company 1 and 2:

days = [1,2,3,4,5,6,7] #days of d week
sales_1 = [160,150,140,145,175,165,180] #sales of company1
sales_2 = [70,90,160,150,140,145,175]  #sales of company2


```python
days = [1,2,3,4,5,6,7] 
sales_1 = [160,150,140,145,175,165,180] 
sales_2 = [70,90,160,150,140,145,175] 
plt.figure(figsize = (12,10))
plt.plot(days,sales_1,linewidth = 5,label = "Sales of Company 1")
plt.plot(days,sales_2,linewidth = 5,label = "Sales of company 2")
plt.xlabel("Days",fontsize = 15)
plt.ylabel("Sales of companies",fontsize  = 15)
plt.legend(shadow = 1,loc = "best")
plt.show()
```

Create a 3 by 3 subplots:
multiple plots

y2 = [10,20,30,40]
y3 = [40,30,20,10]
y4 = [1,2,1,2]

```python
x = [1,2,3,4]
y1 = [4,3,2,1]
y2 = [10,20,30,40]
y3 = [40,30,20,10]
y4 = [1,2,1,2]
y5 = [40,70,90,70]
plt.subplot(3,2,1)
plt.plot(x,y1,"r--")
plt.subplot(3,2,2)
plt.plot(x,y2,"g--")
plt.subplot(3,2,3)
plt.plot(x,y3,"b*--")
plt.subplot(3,2,4)
plt.plot(x,y4,"go--")
plt.subplot(3,2,5)
plt.plot(x,y5,"r*-")
plt.show()
```
